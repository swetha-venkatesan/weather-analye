```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('/content/weather_prediction_dataset.csv')
```

```python
# Preview the data
print(df.shape)
df.head()
```

```python
# Convert 'DATE' to datetime
df['DATE'] = pd.to_datetime(df['DATE'])

# Check missing values
missing = df.isnull().sum()
missing[missing > 0].sort_values(ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot mean temperature of a few cities
cities = ['BASEL_temp_mean', 'BUDAPEST_temp_mean', 'DE_BILT_temp_mean']
df[cities].plot(figsize=(14, 6), title='Mean Temperatures Over Time')
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Highest average temp city
avg_temps = {
    city: df[city].mean() for city in cities
}
sorted(avg_temps.items(), key=lambda x: x[1], reverse=True)
```

```python
# Ensure MONTH column exists (derived from DATE if not already there)
if 'MONTH' not in df.columns:
    df['DATE'] = pd.to_datetime(df['DATE'])
    df['MONTH'] = df['DATE'].dt.month

# Define the 4 EDA features
features = ['temp_mean', 'humidity', 'precipitation', 'sunshine']

# Extract city names that have all 4 features
city_set = set()
for col in df.columns:
    for suffix in features:
        if col.endswith(suffix):
            city = col.replace(f"_{suffix}", "")
            city_set.add(city)

# Plot monthly average trends for each city
for city in sorted(city_set):
    city_cols = [f"{city}_{f}" for f in features]

    # Only plot if all columns exist
    if all(col in df.columns for col in city_cols):
        # Monthly average
        monthly_avg = df.groupby("MONTH")[city_cols].mean()

        # Set Seaborn style
        sns.set(style="whitegrid")
        plt.figure(figsize=(14, 10))

        # Subplots
        for i, feat in enumerate(features):
            plt.subplot(2, 2, i + 1)
            sns.lineplot(x=monthly_avg.index, y=monthly_avg[f"{city}_{feat}"], marker="o")
            plt.title(f'{feat.replace("_", " ").title()} by Month - {city}')
            plt.xlabel("Month")
            plt.ylabel(feat.replace("_", " ").capitalize())

        plt.suptitle(f"📈 Monthly Weather Trends in {city}", fontsize=16)
        plt.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()
```
